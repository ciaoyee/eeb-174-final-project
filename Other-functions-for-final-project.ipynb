```python
#making a dictionary for species and ages, meaning duration of their existence
def GetDuration(filename):
    liz_dict = {} #creates an empty dictionary where the keys and values will go into
    liz_recs = open(filename, "r", encoding = "ISO-8859-15") #will open the file with our data
    for line in liz_recs:
        line = line.replace('\"', '') #removes the extra quotes
        record_elements = line.split(",") #denotes that each record element is separated by a file
        species = record_elements[0] #making species be the keys and taking the names from column 6
        max_age = record_elements[2] #assigning max age to column 15
        min_age = record_elements[3] #assigning min age to column 16
        age = float(max_age) - float(min_age) #calculating the age of the fossil
        liz_dict[species] = age #assigning the age as the values in the dictionary
    return liz_dict
```
